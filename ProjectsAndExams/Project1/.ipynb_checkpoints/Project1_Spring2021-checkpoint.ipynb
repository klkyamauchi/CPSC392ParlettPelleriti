## Data Information

- `reg_to_vote`: 0 if no, 1 if yes.
- `age`: age of the patient in years.
- `hypertension`: 0 if the patient doesn't have hypertension, 1 if the patient has hypertension.
- `heart_disease`: 0 if the patient doesn't have any heart diseases, 1 if the patient has a heart disease.
- `ever_married`: 0 if no, 1 if yes.
- `Residence_type`: 0 for Rural, 1 for Urban.
- `avg_glucose_level`: average glucose level in blood.
- `bmi`: body mass index.
- `smoking_status_smokes`, `smoking_status_formerly`: Whether or not the person smokes, or formerly smoked. If a person has 0's for both these columns, they never smoked.
- `stroke`: 1 if the patient had a stroke or 0 if not.
- `dog_owner`: 0 if no, 1 if yes.
- `er_visits`: number of recorded Emergency Room visits in lifetime.
- `racoons_to_fight`: number of racoons the patient belives they could fight off at once.
- `fast_food_budget_month`: amount (in US dollars) spent on fast food per month.


```python
#imports 

import warnings
warnings.filterwarnings('ignore')


import pandas as pd
import numpy as np
from plotnine import *
from plotnine.data import mtcars

from sklearn.linear_model import LogisticRegression # Logistic Regression Model
from sklearn.preprocessing import StandardScaler #Z-score variables
from sklearn.metrics import mean_squared_error, r2_score #model evaluation
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix

from sklearn.model_selection import KFold # k-fold cv
from sklearn.model_selection import cross_val_predict # cross validation metrics


%matplotlib inline
```

## PART I
Build a logistic regression model to predict whether or not someone had a stroke based on **all** the other variables in the dataset.

1. Count the missing data per column, and remove rows with missing data (if any).
2. Use 10 fold cross validation for your model validation. Store both the train and test accuracies to check for overfitting. **Is the model overfit? How can you tell?**
3. Z-score your continuous variables only.
4. After completing steps 1-3, fit another logistic regression model on ALL of the data (no model validation) using the same predictors as before, and put the coefficients into a dataframe called `coef`.
5. print out a confusion matrix for the model you made in part 4. **What does this confusion matrix tell you about your model? How can you tell?**

```python
#data 
who = pd.read_csv("https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/Proj1.csv")
who.head()
```

```python
# PART I, 1
# Count the missing data per column
missingData = who.isnull().sum(axis=0)
print(missingData)
```

```python
# remove rows with missing data
who = who.dropna()
print(who.isnull().sum(axis=0))
```

```python
# Use 10 fold cross validation for your model validation. 
# Store both the train and test accuracies to check for overfitting. 

# create a K Fold object w/ 10 splits
kf = KFold(n_splits = 10)
```

```python
predictors = ["age", "hypertension", "heart_disease", "ever_married",
              "Residence_type", "avg_glucose_level",
              "bmi","smoking_status_smokes", "smoking_status_formerly", 
              "reg_to_vote", "dog_owner", 
              "raccoons_to_fight", "fast_food_budget_month", 
              "income_in_k", "er_visits"]
X = who[predictors]
y = who[["stroke"]]
```

```python
# create model
lr = LogisticRegression() 

# set up collection for accuracy
mse = []
r2 = []
acc_train = []
acc_test = []
```

```python
for train,test in kf.split(X):
    # Get your train/test for this fold
    X_train = X.iloc[train]
    X_test = X.iloc[test]
    y_train = y.iloc[train] 
    y_test = y.iloc[test]
    
    # Z-score your continuous variables only.
    zScore = StandardScaler()
    zScore.fit(X_train[["age", "avg_glucose_level", "bmi", 
                        "er_visits", "raccoons_to_fight", 
                        "fast_food_budget_month", 
                        "income_in_k", "er_visits"]])
    
    Xz_train = zScore.transform(X_train[["age", "avg_glucose_level", 
                                         "bmi", "er_visits",
                                         "raccoons_to_fight", 
                                         "fast_food_budget_month", 
                                         "income_in_k", "er_visits"]])
    X_train[["age", "avg_glucose_level", "bmi", "er_visits", "raccoons_to_fight",
             "fast_food_budget_month", "income_in_k", "er_visits"]] = Xz_train
    
    Xz_test = zScore.transform(X_test[["age", "avg_glucose_level", "bmi",
                                       "er_visits", "raccoons_to_fight", 
                                       "fast_food_budget_month",
                                       "income_in_k", "er_visits"]])
    X_test[["age", "avg_glucose_level", "bmi", "er_visits",
            "raccoons_to_fight", "fast_food_budget_month", 
            "income_in_k", "er_visits"]] = Xz_test
    
    # model
    model = lr.fit(X_train, y_train)
    
    # record accuracy
    mse.append(mean_squared_error(y_test, model.predict(X_test)))
    r2.append(r2_score(y_test, model.predict(X_test)))
    acc_train.append(accuracy_score(y_train, model.predict(X_train)))
    acc_test.append(accuracy_score(y_test, model.predict(X_test)))
```

```python
print("mean mse: " + str(np.mean(mse)))
print("mean r2: " + str(np.mean(r2)))
print("mean training accuracy: " + str(np.mean(acc_train)))
print("mean testing accuracy: " + str(np.mean(acc_test)))
```

## PART I, 2: Is the model overfit? How can you tell?

To test to see whether the model is overfit to the training data, I calculated the MSE, R2, and accuracies for both the training and testing set. Since we observe a low mse and find that the accuracies of the training set and the testing set are almost identical, we can conclude that our model did a pretty good job at predicting a stroke and that it is *not* overfit.  


```python
# PART 1, 4
# fit another logistic regression model on ALL of the data (no model validation)
myLogit = LogisticRegression() 
myLogit.fit(X, y)

# grab the coefficients and put them into a dataframe called coef
coef = pd.DataFrame({"Coef": myLogit.coef_[0],
                     "Names": predictors})
coef = coef.append({"Coef": myLogit.intercept_[0], 
                    "Names": "intercept"},ignore_index = True)
coef
```

```python
# PART I, 5
# print out a confusion matrix for the model you made in part 4.
plot_confusion_matrix(myLogit, X, y)
```

## PART I, 5:  What does this confusion matrix tell you about your model? How can you tell?

The columns of the confusion matrix represent what our model predicted, and the rows are what the values of the data actually were. On the left hand column, we have the prediction of not having a stroke and the right is the prediction of having a stroke. The top row shows who in real life did not have a stroke and the bottom, those who did not actually have a stroke. The main diagonal shows how often our model was correct. In our upper left hand quadrant, our model correctly predicted that 13,633 people *did not* have a stroke and in the lower right quadrant, we correctly predicted 3 people *did* have a stroke. The opposite diagonal shows us when our predictions were wrong. We predicted that 573 people did not have a stroke but actually did, and predicted that 13 people did have a stroke when they did not actually have one. Observing the ratios of the correct to incorrect predictions given by our confusion matrix, we can see that our model did pretty well.

## Part II, 3. Is your logistic regression model most accurate for people who make less than 30k, between 30-90k, or over 100k? Discuss the potential accuracy *and* ethical implications if your model *were* more accurate for different groups (you can use the full model from part I-4 to check accuracy).

The logistic regression model we built is most accurate for people with incomes less than 30k. While the accuracies are nearly the same across the board, if the differences in accuracies differed more vastly, then we would know that people belonging into that income filter would be under served/represented and could either be more at risk of stroke than the model predicts or be more falsely predicted of being at risk. In cases of stroke, this inaccuracy could mea the loss of actual human lives, which is obviously an immense ethical issue.  

```python
# PART II, 3

predicted = myLogit.predict(X)
who["predicted"] = predicted

# isolate 
less = who.loc[who["income_in_k"] < 30]
between = who.loc[(who["income_in_k"] > 30) 
                  & (who["income_in_k"] < 90)]
over = who.loc[who["income_in_k"] > 100]

# calculate accuracy scores
acc_less = accuracy_score(less["stroke"], 
                          less["predicted"])
acc_between = accuracy_score(between["stroke"],
                             between["predicted"])
acc_over = accuracy_score(over["stroke"], 
                          over["predicted"])

accuracy_df = pd.DataFrame(
    {"Income":["lessThan30k", "between30And90k", "over100k"], 
     "AccuracyScore":[acc_less, acc_between, acc_over]})

accuracy_df.head()
(ggplot(accuracy_df, aes(x="Income", y="AccuracyScore", fill="Income")) 
    + geom_bar(stat = "identity") + theme_minimal()
    + ggtitle("Accuracy of Model Based on Income")
    + labs(y="Accuracy Score"))
```
