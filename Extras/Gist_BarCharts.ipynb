```python
import pandas as pd
import numpy as np

from plotnine import *
```

```python
# Example 1--height of bars is equal to y


#---generate random data---
n = 5
y = np.random.normal(loc = 0, scale = 1, size = n)
x = ["A", "B", "C", "D", "E"]
```

```python
#---build plot---
df = pd.DataFrame({"x":x, "y":y})
ggplot(df, aes(x = "x", y = "y")) + geom_bar(stat = "identity")
```

```python
# Example 2--height of bars is the mean of all rows with A,B, C, D, E respectively

#---generate random data---
n = 100
y = np.random.normal(loc = 0, scale = 1, size = n)
x = np.random.choice(["A", "B", "C", "D", "E"], size = n)
```

```python
#---build plot---
df = pd.DataFrame({"x":x, "y":y})
ggplot(df, aes(x = "x", y = "y")) + stat_summary(fun_data = "mean_sdl", geom = "bar")
```
