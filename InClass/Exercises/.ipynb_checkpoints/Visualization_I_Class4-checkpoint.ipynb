# Together

Let's build a few graphs together!

- show how high blood pressure and age are related
- what is the distribution of platelet counts?
- how many of our participants have a serum sodium > 130?

```python
# data from: https://www.kaggle.com/andrewmvd/heart-failure-clinical-data

heart = pd.read_csv("https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/heart_failure_clinical_records_dataset.csv")

heart.head()
```

```python
(ggplot(heart, aes(x = "factor(high_blood_pressure)", y = "age")) 
 + geom_boxplot())
```

```python
(ggplot(heart, aes(x = "platelets")) 
 + geom_histogram(fill = "white", color = "black"))
```

```python
(ggplot(heart, aes(x = "platelets", group = "sex", fill= "factor(sex)")) 
 + geom_density(color = "black", alpha = 0.3))

#alpha = transparency
```

```python
# how many people have a serum_sodium greater than 130 out of all people
sum(heart.serum_sodium > 130) / heart.shape[0]
```

```python
(ggplot(heart, aes(x = "serum_sodium")) + geom_histogram(fill = "white", color = "black")
+ geom_vline(xintercept = 130, linetype = "dashed", color = "red"))
```
