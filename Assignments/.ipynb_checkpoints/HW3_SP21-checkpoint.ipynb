```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
from plotnine import *

from sklearn.cluster import AgglomerativeClustering

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

from sklearn.cluster import DBSCAN

from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

from sklearn.metrics import silhouette_score

import scipy.cluster.hierarchy as sch
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
# a)

# data
data = pd.read_csv("https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/HW3.csv")

# z-score 
z = StandardScaler()
data[["x","y"]] = z.fit_transform(data)

# ggplot 
ggplot(data, aes(x = "x", y = "y")) + geom_point() + theme_minimal()
```

### b) discuss what patterns you see in the data, and thouroughly discuss the pros/cons of using k-means, Gaussian Mixture Models (with EM), DBSCAN, and Hierarchical Clustering with patterns like the ones you see.

Based on only the scatterplot of the data, it looks like the data can be broken up into three main clusters. A dense sphere on the right, a flag shaped cluster on the left, and another less dense cluster that almost looks like a tetris piece on the bottom. There are also data points scattered across the plot that could be determined as noise. 

**K-Means**: One of the main cons of using K-Means with this data is the assumption that clusters are spherical. Again, looking at the scatterplot we can see that this assumption would probably be broken which would lead us to missing some of the patterns in the clusters. K-Means also forces all the data points into a cluster, so some of the points that look like noise and do not necessarily fit as well into a cluster would forced into one, which would ultimately affect the silhouette score.

**Gaussian Mixture Models (w/ EM)**: With GMM, we do not have the same hard assignment and spherical data assumption that we have with K-Means. Thus, more of the patterns in the clusters could be seen with this model. Like K-Means, GMM also force points into being a part of a cluster, creating the same problem mentioned above. However, the advantage this algorithm has over DBSCAN is that it does not require the clusters to have similar density, making it a good option to use for this data.

**DBSCAN**: I believe DBSCAN would be one of the best options to use on this data as it would be able to recognize both the noise points in the data, as well as the irregular shapes (not spherical or ellipsoid) and patterns in the clusters. However, with DBSCAN we generally like to have clusters with similar densities, which could be a problem when looking at the very dense, spherical cluster in the top right and the bottom, less dense and tetris shaped cluster. DBSCAN also faces trouble with data that may have overlapping or touching clusters. This doesn't seem like it would be too much of a problem for this data set as each cluster seems to have a decent amount of separation from each other. 

**Hierarchicial Clustering**: One of the advantages of HAC is the amount of variation it allows in how we measure distance between data points as well as how it models the relationships between our clusters and how similar they are with each other. A problem we may face is that we cannot un-merge the clusters we have created, once we have created them. For example, for points in the lower left region of plot that could fit into the flag shape cluster, we could find that if we iteratively went through the models and reclustered, the point may actually fit better into the tetris shaped cluster. However, we would have to settle for the suboptimal first clustering that we identified that point in initially since we are unable to unmerge the clusters. We also know that hierarchicial clustering does not work as well on a very large data set. 

### c) choose which algorithm you think will perform the *best* on this data, and the algorithm you think will perform the *worst* on the data. **Briefly explain why you think each algo you chose is the best/worst** (there may be overlap from part b, but please answer this separately, despite repetition).

I think that the **best** algorithm to use on this data would be **GM** since it takes into account the irregular shape of the points, and does not assume a spherical shape in its clusterings. This allows us to consider the irregular patterns of the three clusters. ### d) using the TWO algorithms you chose in part c, cluster the data and calculate the silhouette scores. If you had to choose any hyperparameter values (including but not limited to: k, distance metric, linkage critera, eps, min_samples...etc), state how you chose them. Which algorithm did better according to the silhouette scores and plots from part e?

According to the silhouette scores, The GM model did much better. It had a silhouette score of 0.71 while the HAC model had a silhouette score of 0.61. I chose the linkage of the HAC model based on the centroids to not include the outlier points in the data, also as this is one of the more standard linkages used for this model. For the GM model and the HAC model, I found it quite obvious just by looking at the regularly plotted points that there were 3 clusters, so that is what I chose for n_components as well. I double checked that this would be the best clustering using the dendrogram for the data as well. 



```python
# d)

# build the model with the number of clusters we want and fit the data
EM = GaussianMixture(n_components = 3)
EM.fit(data)

# get the most likely cluster for each data point
cluster = EM.predict(data)


print("GM Silhouette Score: ", silhouette_score(data, cluster))

# adding cluster membership to the data frame since ggplot needs it
data["cluster"] = cluster
```

```python
X = data[["x","y"]]

hac = AgglomerativeClustering(affinity = "euclidean",
                             linkage = "ward")

hac.fit(X)

dendro = sch.dendrogram(sch.linkage(X, method='ward'))
```

```python
hac = AgglomerativeClustering(n_clusters = 3,
                              affinity = "euclidean",
                             linkage = "ward")
hac.fit(X)

# silhouette score
groups = hac.labels_
data["hac"] = groups

print("HAC Silhouette: " + str(silhouette_score(X,groups)))
```

```python
# EM Plot
print(ggplot(data, aes(x = "x", y = "y",
                    color = "factor(cluster)")) +
             geom_point() +
             theme_minimal() +
             scale_color_discrete(name = "Cluster Assignment") +
             theme(panel_grid_major = element_blank()) + 
             labs(title = "Gaussian Mixtures"))
```
